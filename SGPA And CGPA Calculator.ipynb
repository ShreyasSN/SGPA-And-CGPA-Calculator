# SGPA and CGPA Calculator

```python
import numpy as np
import matplotlib.pyplot as plt
```

### Grade Points Weightage

```python
S = 10
A = 9
B = 8
C = 7
D = 6
E = 5
F = 0
P = 0
R = 4
```

### Grades scored and respective credits for the courses in the each semester.

```python
# [[Course credit, Grade sored], [next course grades], [and so on]]

sem1 = [[2,B],[3,B],[3,C],[2,S],[1,A],[3,A]]
sem2 = [[2,B],[3,B],[3,S],[2,S],[1,B],[3,A]]
sem3 = [[2,B],[3,B],[3,C],[2,C],[1,B],[3,A]] 
sem4 = [[2,B],[3,B],[3,C],[2,S],[1,B],[3,A]]
sem5 = [[2,B],[3,A],[3,C],[2,S],[1,B],[3,S]]
sem6 = [[2,B],[3,S],[3,C],[2,B],[1,B],[3,A]]
sem7 = [[2,C],[3,A],[3,C],[2,S],[1,B],[3,D]]
sem8 = [[2,S],[3,S],[3,C],[2,S],[1,B],[3,S]]
```

```python
# Please specify the semesters to be included or excluded for SGPA calculation.

semesters = [sem1, sem2, sem3, sem4, sem5,sem6,sem7,sem8]
no_of_semesters = len(semesters)
```

```python
def sgpa(credits):
    sum = 0
    credit = 0
    rows = len(credits)
    for i in range (rows):
        sum += credits[i][0]*credits[i][1]
        credit += credits[i][0]
    return round(sum/credit, 2)
```

## SGPA Calculation

```python
sgpa_values = []
for i in range (no_of_semesters):
    result = 0
    sgpa_value = sgpa(semesters[i])
    sgpa_values.append(sgpa_value)
    result = 'SEM {} SGPA : {}'.format(i+1,sgpa_value)
    print(result)
```

## SGPA Graph

```python
#SGPA VS Semesters graph to track SGPA
import seaborn as sns
x_values = list(range(1, no_of_semesters+1))
sns.set_style("darkgrid")
plt.plot(x_values, sgpa_values, 'go-') 
plt.xlabel('Number Of Semesters')
plt.ylabel('SGPA Obtained')
plt.title('SGPA vs SEM')
plt.show()
```

## CGPA Calculation

```python
# Please specify the semesters to be included or excluded for CGPA calculation.

total_semester = np.concatenate((sem1, sem2,sem3,sem4,sem5,sem6,sem7,sem8))
cgpa = 'CGPA : {}'.format(sgpa(total_semester))
print(cgpa)
```
